#### Define function to check null values

```python
def check_null(df):
    na = df.isna().sum()
    na = pd.DataFrame(na, columns=['null'])

    # calculate percentage of null values
    na['percentage'] = na['null'] / df.shape[0] * 100
    na = na.sort_values(by='percentage', ascending=False)
    return na

na = check_null(df)
text_na = check_null(text_df)
```

## Drop columns

```python
# normal df
col_drop = na[na['percentage'] > 70].index
df = df.drop(col_drop, axis=1)
# text df
col_drop = text_na[text_na['percentage'] > 70].index
text_df = text_df.drop(col_drop, axis=1)
df.head()
```

## Convert to categorical

```python
# function to convert numeric to categorical variables
def convert_categorical(df):
    for name in df:
        # calculate ratio of unique values to total values
        unique = df[name].nunique()
        total = df.shape[0]
        if unique/total < 0.05:
            # print(name)
            df[name] = df[name].astype('category')
    return df

df = convert_categorical(df)
eda_df = convert_categorical(eda_df)
df.info()
```

#### Set numeric variables to float

```python
numeric = ['nperpcap', 'nkill', 'nkillus', 'nkillter', 'nwound', 'nwoundus', 'nwoundte']
df[numeric] = df[numeric].astype(float)
eda_df[numeric] = eda_df[numeric].astype(float)
```

## Check for outliers in numeric columns

```python
def count_outlier(df):
    rows = df.shape
    outlier = {}
    for col in df:
        if (df[col].dtype == 'category'):
            continue
        q1 = np.quantile(df[col], 0.25)
        q3 = np.quantile(df[col], 0.75)
        iqr = q3-q1
        upper = q3+1.5*iqr
        lower = q1-1.5*iqr
        count = sum((df[col] > upper) | (df[col] < lower))
        outlier[col] = count / rows[0] * 100
        outlier_df = pd.DataFrame.from_dict(outlier, orient='index', columns=['outliers'])
    return outlier_df
        
```

```python
outlier = count_outlier(df)
outlier
```

there are no outliers

## Impute NULL values

```python
numeric_cols = df.select_dtypes(include='number').columns
categoric_cols = df.select_dtypes(include='category').columns
```

```python
df.shape
```

```python
def impute_null(df):
    na_col = df.isna().any() # check if column containes null value
    # impute missing value for numeric cols
    imputer = SimpleImputer(strategy='mean')
    for col in numeric_cols:
        if na_col[col]:
            col_data = df[[col]]
            imputed_col = imputer.fit_transform(col_data)
            # replace original column with imputed values
            imputed_col = pd.DataFrame(imputed_col)
            df[col] = imputed_col

    # impute missing value for categorical cols
    imputer_categorical = SimpleImputer(strategy='most_frequent')
    for col in categoric_cols:
        if na_col[col]:
            col_data = df[[col]]
            imputed_col = imputer_categorical.fit_transform(col_data)
            # replace original column with imputed values
            imputed_col = pd.DataFrame(imputed_col)
            df[col] = imputed_col

    return df

df = impute_null(df)
eda_df = impute_null(eda_df)
```
